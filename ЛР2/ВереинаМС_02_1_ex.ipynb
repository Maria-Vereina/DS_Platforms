## 2.1.3: Objects

Создайте класс `Pet` с параметрами `name, pronoun, animal,` и `pet_response`.

Добавьте метод `pet()`, который печатает ответ, состоящий из элементов, например:

` Лейла — кошка. Если ее погладить, Лейла мурчит».`
` Скаут — собака. Если его погладить, Скаут виляет хвостом.`

1) Создам класс Pet и добавлю метод pet(), который печатает ответ

````python
class Pet:

      def __init__(self, name, pronoun, animal, pet_response):
        self.name = name
        self.pronoun = pronoun
        self.animal = animal
        self.pet_response = pet_response

      def pet(self):
        print(f"{self.name} - {self.animal}. Если {self.pronoun} погладить, {self.name} {self.pet_response}.")
````

2) Вызову функцию. Вот несколько примеров вызовов:

````python
layla = Pet("Лейла", "ее", "кошка", "мурчит")
scout = Pet("Скаут", "его", "собака", "виляет хвостом")

layla.pet()
scout.pet()
````

4) Обновить данные, чтобы вместо количества чемпионатов отражалось количество побед.

5.2) Обновите данные, чтобы они содержали строку суммы в долларах, включая знак $. Например, для Локомотив это будет соответствовать:

  ```
  Тренер Юрий Сёмин  Локомотив $ 600000
  ```

````python
#функция для добавления знака $

def toDollars(amount):
    return "$ " + str(amount)

#создание нового датафрейма для необходимых преобразований

cols = ['Тренер', 'Всего']
new_df = df2.loc[:, cols].copy()

#создаем столбец, чтобы на первом месте выводилось "Тренер", как в примере

new_df['Должность'] = 'Тренер'
new_df.set_index('Должность', inplace=True)

#используем функцию на весь столбец датафрейма

new_df['Прибыль'] = new_df['Всего'].apply(toDollars)

#удаляем ненужный столбец, чтобы данные выводились как в примере

new_df.drop(labels=new_df.columns[1], axis=1, inplace=True)

# Вывод нового датафрейма

print(new_df)
````
