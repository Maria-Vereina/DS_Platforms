## 2.2.1: Reading and Writing Data

Файл [grades.csv](grades.csv) представляет собой файл с именами учащихся и буквенными оценками:

```
Alice; A
Bob; B
Robert; A
Richard; C
```

Записать данные из файла в массив. Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

Подсказка: функция [strip()](https://docs.python.org/3/library/stdtypes.html#str.strip) удаляет конечные пробелы из строки.

Запишите этот файл в новый файл `grades_gpa.csv`.

1) Импортируем исходный файл

````python
from google.colab import files

grades = files.upload()
````

2) Записать данные из файла в массив. Добавить средний балл учащегося в строку учащегося и записать этот файл в новый файл grades_gpa.csv.

````python
# импортировать библиотеку csv
import csv

gpas = {"A":4, "B":3, "C":2, "D":1}

# открытие файла и добавление строк в виде массивов,
# подготовка файла для записи

with open('grades.csv') as csvfile:
  with open('grades_gpa.scv', 'w') as csv_writer:
    filewriter = csv.writer(csv_writer, delimiter=';')

    filereader = csv.reader(csvfile, delimiter=';')

# строки здесь представляют собой массив

    for row in filereader:
        result = list(map(lambda x: x.strip(), row))
        result.append(gpas[result[1]])
        filewriter.writerow(result)
        print(result)
````
